```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('carprices.csv')
df.head()
```

```python
%matplotlib inline
plt.xlabel('Mileage', fontsize = 20)
plt.ylabel('Sell Price($)', fontsize = 20)
plt.scatter(df['Mileage'], df['Sell Price($)'], color = 'red', marker = '+')
```

```python
%matplotlib inline
plt.xlabel('Age(yrs)', fontsize = 20)
plt.ylabel('Sell Price($)', fontsize = 20)
plt.scatter(df['Age(yrs)'], df['Sell Price($)'], color = 'blue', marker = '*')
```

```python
X = df[['Mileage', 'Age(yrs)']]
y = df['Sell Price($)']
```

```python
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
X_train
```

```python
clf = LinearRegression()
```

```python
clf.fit(X_train, y_train)
```

```python
clf.predict(X_test)
```

```python
y_test
```

```python
clf.score(X_test, y_test)
```
